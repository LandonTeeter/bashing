```python
dhfr_variants = variants[variants['GENE'].str.contains('PF3D7_0417200')]
display(dhfr_variants.head())
```

```python
dhfr_missense_variants = dhfr_variants[dhfr_variants['EFFECT'] == 'missense_variant']
display(dhfr_missense_variants.head())
```

```python
import altair as alt

dhfr_108_variants = dhfr_missense_variants[dhfr_missense_variants['AA_POS'] == 108]
display(dhfr_108_variants.head())
```

```python
# Normalize Sample names by removing '.fq' suffix for consistency
dhfr_108_variants_normalized = dhfr_108_variants.copy()
dhfr_108_variants_normalized['Sample'] = dhfr_108_variants_normalized['Sample'].str.replace('.fq', '', regex=False)

# Create a list to store data for the scatter plot, including reference alleles
scatter_plot_rows = []

# Iterate through each unique sample to ensure all four samples are represented
# This also handles cases where a sample might not have the specific variant
# but we still want to show its reference state.
all_samples = ['ERR636434', 'ERR636028', 'ERR042232', 'ERR042228']

# Define the reference position and alleles
reference_chrom = 'NC_004318.2'
reference_pos = 748410
original_ref_allele = 'G'
reference_hgvs_p = 'p.Ser108'

for sample in all_samples:
    sample_variants = dhfr_108_variants_normalized[dhfr_108_variants_normalized['Sample'] == sample]

    if not sample_variants.empty:
        # If the sample has a variant at this position
        for index, row in sample_variants.iterrows():
            scatter_plot_rows.append({
                'Sample': row['Sample'],
                'POS': row['POS'],
                'AF': row['AF'],
                'CHROM': row['CHROM'],
                'REF': row['REF'],
                'ALT': row['ALT'],
                'HGVS_P': row['HGVS_P']
            })
    else:
        # If the sample does not have a variant, assume reference allele with AF=0
        scatter_plot_rows.append({
            'Sample': sample,
            'POS': reference_pos,
            'AF': 0.0,
            'CHROM': reference_chrom,
            'REF': original_ref_allele,
            'ALT': original_ref_allele,
            'HGVS_P': reference_hgvs_p # Assuming 'Ser108' as the reference amino acid
        })

scatter_plot_data = pd.DataFrame(scatter_plot_rows)
display(scatter_plot_data)

# Create the scatter plot using Altair, with color encoding for AF
chart = alt.Chart(scatter_plot_data).mark_point().encode(
    x=alt.X('POS', title='Variant Coordinate (POS)'),
    y=alt.Y('Sample', title='Sample Accession'),
    fill=alt.Color('AF', title='Alternative Allele Frequency', legend=alt.Legend(format='.0%'), scale=alt.Scale(scheme='redblue')),
    stroke=alt.value('black'),
    strokeWidth=alt.value(1),
    tooltip=[
        alt.Tooltip('Sample'),
        alt.Tooltip('CHROM', title='Chromosome'),
        alt.Tooltip('POS', title='Position'),
        alt.Tooltip('REF', title='Reference Allele'),
        alt.Tooltip('ALT', title='Alternative Allele'),
        alt.Tooltip('AF', title='Allele Frequency', format='.2%'),
        alt.Tooltip('HGVS_P', title='Protein Change')
    ]
).properties(
    title='Allele Frequencies of dhfr gene variants at AA Position 108'
)

chart.display()
```

## Aggregate Resistance Allele Frequencies by Location

### Subtask:
Consolidate the allele frequency data from the samples for each location. For each country, identify the resistance allele frequency (the non-zero AF from the two samples). Then, prepare a DataFrame that contains the resistance allele frequency and the corresponding non-resistance frequency (1 - resistance AF) for each location, along with its coordinates. This will be used to create the pie chart slices.


**Reasoning**:
To consolidate the allele frequency data, I will group the `scatter_plot_data_merged` DataFrame by 'Country' and find the maximum 'AF' for each country, representing the resistance allele frequency. Then, I will calculate the non-resistance allele frequency.



```python
resistance_af_by_country = scatter_plot_data_merged.groupby('Country')['AF'].max().reset_index()
resistance_af_by_country = resistance_af_by_country.rename(columns={'AF': 'resistance_af'})
resistance_af_by_country['non_resistance_af'] = 1 - resistance_af_by_country['resistance_af']
display(resistance_af_by_country.head())
```

**Reasoning**:
I will now melt the `resistance_af_by_country` DataFrame to convert it into a long format, which is more suitable for creating pie charts in Altair. This will create 'type' and 'frequency' columns. After that, I will merge this melted DataFrame with `location_df` to incorporate the geographical coordinates for each country.



```python
pie_chart_data_long = resistance_af_by_country.melt(id_vars=['Country'], var_name='type', value_name='frequency')

# Merge with location_df to add geographical coordinates
pie_chart_data_final = pd.merge(pie_chart_data_long, location_df, on='Country')

display(pie_chart_data_final.head())
```
